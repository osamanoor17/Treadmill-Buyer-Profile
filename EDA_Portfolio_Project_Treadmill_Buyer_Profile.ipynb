***Outlier Detection:***

```python
# Check for the outliers by using the IQR method.

# Select only numerical columns
numeric_data = data.select_dtypes(include=['int64', 'float64'])

# Detecting outliers using IQR
Q1 = numeric_data.quantile(0.25)  # First quartile (25th percentile)
Q3 = numeric_data.quantile(0.75)  # Third quartile (75th percentile)
IQR = Q3 - Q1  # Interquartile range

# Define outlier boundaries
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Find outliers
outliers = ((numeric_data < lower_bound) | (numeric_data > upper_bound))

# Display the count of outliers per column
outlier_counts = outliers.sum()
print("Outlier Counts per Column:\n", outlier_counts)
```

```python
#Percentage of customers with Age between 20s and 30s among all customers
data['Product'] = data['Product'].astype(str)

# Define the selected products
selected_products = ['KP281', 'KP481', 'KP781']

# Filter customers who purchased the selected products
customers_with_selected_products = data[data['Product'].isin(selected_products)]

# Filter customers in the age range 20-39 (20s and 30s)
customers_in_age_range = data[(data['Age'] >= 20) & (data['Age'] < 40)]

# Calculate percentage of customers in age range
percentage_age_ranges = (len(customers_in_age_range) / len(data)) * 100

print(f"Percentage of customers with Age between 20s and 30s among all customers: {percentage_age_ranges:.2f}%")
```

```python
# Product – Income
# Percentage of a low-income customer purchasing a treadmill
product_income_counts = data.groupby(['Product', 'Income']).size().unstack()
print("Product-Income Frequency Table:")
print(product_income_counts)
```

```python
# Percentage of a high-income customer purchasing KP781
# Ensure 'High' exists in the DataFrame

# Define high-income threshold (e.g., above $70,000)
high_income_threshold = 70000

# Select only columns (income levels) above the threshold
high_income_columns = [col for col in product_income_counts.columns if col > high_income_threshold]

if 'KP781' in product_income_counts.index:
    # Sum purchases for KP781 among high-income customers
    kp781_high_income_purchases = product_income_counts.loc['KP781', high_income_columns].sum()

    # Sum all high-income customer purchases across all products
    total_high_income_purchases = product_income_counts[high_income_columns].sum().sum()

    # Calculate percentage
    percentage_high_income_kp781 = (kp781_high_income_purchases / total_high_income_purchases) * 100

    print(f"Percentage of high-income customers purchasing KP781 treadmill: {percentage_high_income_kp781:.2f}%")
else:
    print("Error: KP781 not found in dataset.")
```

```python
# Percentage of customer with high-income salary buying treadmill given that Product is KP781

# Defining a high-income threshold..
high_income_threshold = 70000

# Select only income levels above the threshold
high_income_columns = [col for col in product_income_counts.columns if col > high_income_threshold]

# Check if KP781 exists in dataset
if 'KP781' in product_income_counts.index:
    # Total customers who purchased KP781 with high income
    high_income_kp781_customers = product_income_counts.loc['KP781', high_income_columns].sum()

    # Total high-income customers across all products
    total_high_income_customers = product_income_counts[high_income_columns].sum().sum()

    # Calculate percentage
    percentage_high_income_kp781 = (high_income_kp781_customers / total_high_income_customers) * 100

    print(f"Percentage of high-income customers purchasing KP781 treadmill: {percentage_high_income_kp781:.2f}%")
else:
    print("Error: KP781 not found in dataset.")
```

***Actionable Insights & Recommendations***

1. Customer Demographics & Buying Behavior
**Insights:**

- Majority of the customers belong to the age group of 20s and 30s.
- The purchase rate of KP281, KP481, and KP781 is too lower compared to other products.
- High-income customers (salary > 70,000) contribute a specific percentage of total treadmill purchases, but their engagement with KP781 is relatively low.

**Recommendations:**
- Targeted Marketing: Since customers in their 20s and 30s are more active buyers, they were not focused with the advertising campaigns rather than focusing on fitness benefits for this age group.
- Product Bundling: To boost sales of KP281, KP481, and KP781, introduce discounts, EMI options, or bundle deals with accessories.
- Income-Based Offers: Provide premium plans or discounts for high-income customers to increase engagement with KP781.

2. Correlation Analysis & Customer Preferences
**Insights:**

- Strong correlation observed between age and product preference (younger customers prefer budget-friendly options).
- Weak correlation between income and purchase (high-income customers do not necessarily buy more expensive treadmills).

**Recommendations:**
- Age-Based Personalization: Introduce personalized product suggestions based on customer demographics.
- Luxury Features for High-Income Groups: Since high-income customers are not buying KP781 frequently, enhance premium features (e.g., smart fitness tracking, extended warranty, or VIP membership).

3. Outlier Analysis & Data Quality

**Insights:**
- Presence of outliers in income and age suggests potential data entry errors or niche customer segments.
- Some customers have exceptionally high salaries but do not purchase high-end treadmills.

**Recommendations:**
- Data Cleaning: Ensure consistent data formatting and remove anomalies affecting analysis.
- Special Offers for Niche Segments: If outliers represent niche segments (e.g., older or ultra-high-income individuals), design exclusive marketing strategies for them.

4. Sales Performance & Product Positioning
**Insights:**

- KP281 and KP481 have higher sales volume, but KP781 apparently had less.
Certain income groups prefer specific treadmill models.
Recommendations:
- Reposition of KP781: Promote KP781 using premium branding strategies such as influencer marketing or high-end fitness endorsements.
- Competitive Pricing Adjustments: Analyze whether price adjustments or installment-based purchases can improve high-ticket product sales.
